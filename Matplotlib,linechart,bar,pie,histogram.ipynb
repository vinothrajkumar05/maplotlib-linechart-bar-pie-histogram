```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([20,35,50,90])
mylabel=["apple","orange","kiwi","blueberry"]
plt.pie(x,labels=mylabel)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([20,35,50,90])
mylabel=["apple","orange","kiwi","blueberry"]
seperate=[0,0,0,0.3]
plt.pie(x,labels=mylabel,startangle=90,explode=seperate)
plt.show()
```
